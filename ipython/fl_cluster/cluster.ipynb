```python
%pylab inline
import pylab
from __future__ import print_function
import itertools as it
import operator as op
from collections import defaultdict, Counter
from sklearn.manifold import TSNE
import json

from gensim import corpora, models
from gensim.models import Word2Vec


import chain_converter as chain_converter


import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)


train_data = 'data'
```

##### Loading chain converter

```python
with open('chains', 'r') as f:
    converter = chain_converter.ChainsHier(json.load(f))
clusters = converter.hier.keys()[:-2]
print(clusters)
```

```python
def split_string_features(ns_values):
    ns_values_dict = {}
    for fvalue in ns_values:
        if fvalue.find(':') != -1:
            fname, fvalue = fvalue.split(':')
            fvalue = float(fvalue)
            fname = fname
            ns_values_dict[fname] = fvalue
        else:
            ns_values_dict[fvalue] = 1
    return ns_values_dict


def split_vw_string(s, separator='|'):
    s = s.decode('utf8')
    classvar, _, ns = (part.strip() for part in s.partition(separator))
    classvar, ticket_id = classvar.split()
    features = {}
    namespaces = (part.strip() for part in ns.split(separator))
    for ns in namespaces:
        ns_name, _, ns_values = ns.partition(' ')
        #ns_values_dict = features.get(ns_name, {})
        ns_values_dict = split_string_features(ns_values.split())
        features[ns_name] = ns_values_dict
        
    tokens = features['tokens']
    meta = features.get('meta', {})
    for k,v in meta.items():
        if k.startswith('topic') or k.startswith('another'):
            tokens[k] = v
    return classvar, ticket_id, tokens


def reader(f):
    for row in f:
        classvar, ticket_id, tokens = split_vw_string(row)
        if len(tokens) > 15 and len(tokens) < 200:
            yield tokens

            
def full_reader(f):        
    for row in f:
        classvar, ticket_id, tokens = split_vw_string(row)
        if len(tokens) > 15 and len(tokens) < 200:
            yield classvar, ticket_id, tokens

        
class Corpora(object):
    def __iter__(self):
        with open(train_data, 'r') as rf:
            for doc in reader(rf):
                #yield dictionary.doc2bow(doc)
                yield doc
              
```

##### Learning word2vec model from full data file

```python
corpus = Corpora()
model = Word2Vec(corpus, size=100, min_count=5)
```

##### Use word2vec model to data

```python
cluster_limit = 500
cluster_vectors = Counter()
word2vec_result = 'word2vec'


with open(train_data, 'r') as rf, \
    open(word2vec_result, 'w') as wf:
    for classvar, ticket_id, tokens in full_reader(rf):
        cluster = converter.get_cluster(classvar, vw=True)
        cluster_vectors[cluster] += 1
        if cluster_vectors[cluster] < cluster_limit and cluster is not None:
            tokens = [k for k, v in tokens.items() if k in model]
            tokens =  sum(model[f] for f in tokens) / len(tokens)
            wf.write('%s\t%s\t%s\n' % (classvar, ticket_id, ' '.join(str(f) for f in tokens)))

labels = []
data = []
tickets = []
with open(word2vec_result, 'r') as rf:
    for line in rf:
        classvar, _, msg = line.partition('\t')
        ticket_id, _, msg = msg.partition('\t')
        labels.append(classvar)
        data.append(msg.split())
        tickets.append(ticket_id)

limit = 10000
subset_data = data[:limit]
subset_labels = labels[:limit]
subset_tickets = tickets[:limit]

subset_body = {}
with open('body', 'r') as f:
    for row in f:
        classvar, ticket_id, body = row.split(' ', 2)
        if ticket_id in subset_tickets:
            subset_body[ticket_id] = body


print('Data for tsne has %s exampels' % len(subset_data))
print('Body for visualize %s exampels' % len(subset_body))
```

```python
#Traininng TSNE
tsne = TSNE(n_components=2, random_state=0)
result = tsne.fit_transform(subset_data)
```

```python
#plotting clusters
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.cm as cmx
from matplotlib.markers import MarkerStyle

def get_cmap(N):
    '''Returns a function that maps each index in 0, 1, ... N-1 to a distinct 
    RGB color.'''
    color_norm  = matplotlib.colors.Normalize(vmin=0, vmax=N-1)
    scalar_map = cmx.ScalarMappable(norm=color_norm, cmap='hsv') 
    def map_index_to_rgb_color(index):
        return scalar_map.to_rgba(index)
    return map_index_to_rgb_color




num_colors = len(clusters)
print('Num chains: %s' % num_colors)


markers = MarkerStyle().filled_markers
colors = get_cmap(num_colors)

plt.figure(figsize=(16, 16))
scatters = []
legends = []
for i, cluster in enumerate(clusters):
    spam = [point for point, label in it.izip(result, subset_labels) if cluster == converter.get_cluster(label, True)]
    s = plt.scatter(map(op.itemgetter(0), spam), map(op.itemgetter(1), spam), c=colors(i), marker=markers[i], s=30)
    scatters.append(s)
    legends.append('cluster %s' % cluster)
plt.legend(scatters, legends,
           scatterpoints=1,
           loc='lower left',
           ncol=3,
           fontsize=16)
plt.show()
```

```python
plotly_data = []
with open('to_plot', 'w') as f:
    for point, label, ticket in it.izip(result, subset_labels, subset_tickets):
        cluster = converter.get_cluster(label, True)
        x, y = point
        body = subset_body[ticket]
        body = body.replace('\\n', '\n')
        body = body.replace('\n', '<br>')
        if cluster is not None:
            info = '%s %s %s\n%s' % (cluster, converter.chains_from_vw[label], ticket, body)
            s = '%s\t%s\t%s' % (x, y, info)
            plotly_data.append([x, y, cluster, converter.chains_from_vw[label], ticket, body])
            print(s, file=f)
            
```
